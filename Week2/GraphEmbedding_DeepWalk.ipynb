# Generating 3D Embeddings of a random Graph

## Generating a random graph

### Random graph using erdos_renyi_grpah function

```python
import networkx as nx

n = 60 #No. of Nodes
p = 0.5 #Probability of connection 
g = nx.erdos_renyi_graph(n,p)

#Printing the Nodes
print(g.nodes)
```

```python
#Printin the edges
print(g.edges)
```

### Drawing the Graph

```python
nx.draw(g, node_size = n)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from karateclub.node_embedding.neighbourhood.deepwalk import DeepWalk
```

### Generating a 3 Dimensional embedding

```python
deepwalk = DeepWalk(dimensions = 3)
deepwalk.fit(g)
embedding = deepwalk.get_embedding()
plt.scatter(embedding[:, 0], embedding[:, 1])
plt.show()
```

### Plotting the embedding in a 3D graph

```python
fig = plt.figure(figsize=(15,8))
ax =fig.add_subplot(projection = "3d")
ax.scatter(embedding[:, 0], embedding[:, 1], embedding[:, 2])
plt.show()
```
