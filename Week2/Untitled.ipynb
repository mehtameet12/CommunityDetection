# Creating a random graph using networkx and erdos_renyi_grpah function

```python
import networkx as nx

n = 60 #No. of Nodes
p = 0.5 #Probability of connection 
g = nx.erdos_renyi_graph(n,p)

print(g.nodes)
```

```python
## Printin the edges

print(g.edges)
```

```python
## Drawing the Graph
nx.draw(g, node_size = 6)
```
